```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_context("paper", font_scale = 2)
```

### Style Guides

The default style is of matplotlib

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Setting seaborn default style

```python
sns.set() # Setting seaborn default
```

```python
sns.displot(data=tips,
            x="total_bill",
            kind="hist",
            bins = 20,
            aspect = 1.5)
```

```python
sns.set_style("white") # Setting white
```

```python
sns.displot(data=tips, x="total_bill", kind="hist", bins = 20, aspect = 1.5)
```

```python
["darkgrid", "whitegrid", "dark", "white", "ticks"] # Avilable style options
```

```python
for style in ["darkgrid", "ticks"]:
    sns.set_style(style)
    sns.displot(data=tips,
                x="total_bill",
                kind="hist",
                bins = 20,
                aspect = 1.5)
```

### Removing axes lines

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.stripplot(x = "day",
              y = "tip",
              data = tips,
              jitter = True,
              ax = ax,
              s = 8)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.stripplot(x = "day",
              y = "tip",
              data = tips,
              jitter = True,
              ax = ax,
              s = 8)
sns.despine(right = True)

plt.show()
```

## Color in seaborn

### Assigning matplotlib color code

```python
sns.set(color_codes = True)
sns.displot(data=tips,
            x="total_bill",
            color = "red",
            kind="hist",
            bins = 20,
            aspect = 1.5)
```

#### Let's add two colors from magma palette

```python
magma_2 = sns.palplot(sns.color_palette("magma", n_colors=2))
```

```python
magma_2 = sns.color_palette("magma", n_colors=2)

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x = "total_bill",
              y = "tip",
              data = tips,
              ax = ax,
              hue = "sex",
              palette = magma_2,
              s = 50)
sns.despine(right = True)
plt.show()
```

### Setting a default palette

```python
sns.set_palette(
    palette = "magma",
    n_colors = 4,
)
```

```python

fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(x = "day",
              data = tips,
              ax = ax)
plt.show()
```
